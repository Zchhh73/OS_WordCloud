```python
#coding:utf-8
from wordcloud import WordCloud
import cv2
import jieba
```

```python
with open('os.txt','rb') as f:
    text = f.read()
cut_text =" ".join(jieba.cut(text))
color_mask = cv2.imread('OS.jpg')

cloud = WordCloud(font_path='C:\Windows\Fonts\STXINWEI.TTF',background_color='white',mask=color_mask,max_words=500,max_font_size=120)

wcloud = cloud.generate(cut_text)
wcloud.to_file('cloud_os.jpg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(wcloud,interpolation="bilinear")
plt.axis('off')
plt.show()
```
